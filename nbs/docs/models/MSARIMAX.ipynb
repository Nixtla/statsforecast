# MSARIMAX Model

> Fixed-order multiple-seasonal SARIMAX model for time series with more than one seasonal pattern.

`MSARIMAX` expands AR, differencing, and MA orders across multiple seasonal lag levels. It supports exogenous regressors, prediction intervals, and in-sample fitted values through the standard StatsForecast model interface.

## Example

The example below simulates hourly data with daily and weekly seasonalities, then fits a fixed multiple-seasonal SARIMAX specification.

```python
import numpy as np
import pandas as pd

from statsforecast import StatsForecast
from statsforecast.models import MSARIMAX


def simulate_two_seasonal(n=240, seed=42):
    rng = np.random.default_rng(seed)
    t = np.arange(n)
    daily = 3.0 * np.sin(2 * np.pi * t / 24)
    weekly = 1.8 * np.cos(2 * np.pi * t / 168)
    trend = 0.01 * t
    noise = rng.normal(scale=0.35, size=n)
    return 20 + trend + daily + weekly + noise


y = simulate_two_seasonal()
df = pd.DataFrame(
    {
        "unique_id": "series_1",
        "ds": pd.date_range("2024-01-01", periods=len(y), freq="h"),
        "y": y,
    }
)

model = MSARIMAX(
    lags=[1, 24, 168],
    ar_order=[0, 0, 1],
    i_order=[0, 1, 0],
    ma_order=[0, 0, 0],
)
sf = StatsForecast(models=[model], freq="h")
sf.forecast(df=df, h=48, level=[80, 95]).head()
```
